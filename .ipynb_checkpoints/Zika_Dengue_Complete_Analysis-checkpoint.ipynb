# Cross-Flavivirus Convergent Host Response
## GSE110496 — Single-Cell RNA-seq Pipeline (DENV + ZIKV, Huh7 Hepatocytes)

**Dataset:** Zanini et al., eLife 2018 — single-cell RNA-seq of Huh7 cells infected with
DENV-2 or ZIKV at 4h, 12h, 24h, 48h post-infection.

**Central hypothesis:** DENV and ZIKV exploit a conserved flavivirus host regulatory network —
the same biological programs, pathways, and host factors are activated or suppressed by both viruses.

**Three evidence layers:**
1. Literature-derived DENV-downregulated miRNAs (91 total; 55 cross-flavivirus, 36 DENV-only control)
2. Transcriptomic data — primary: GSE110496; validation: GSE118305, GSE94892, GSE78711
3. Literature-curated proviral/antiviral host factor gene lists

---

## HOW TO USE THIS NOTEBOOK

**`Kernel → Restart & Run All`** — runs the complete pipeline from scratch.

**Checkpoint system:** Every step saves a JSON checkpoint in `checkpoints/`.  
If a step was already completed, it will be **skipped automatically** — no repeated computation.
To force re-run a step, delete its checkpoint file: `checkpoints/stepXX_checkpoint.json`

**Conda environment:** Activate `scipy` before launching Jupyter:
```bash
conda activate scipy
jupyter notebook
```

## Cell 0 — Shared Setup
Run this first. Sets paths, defines the checkpoint runner, and verifies the environment.

````python
import importlib.util
import os
import sys
import json
from pathlib import Path

# ── Project root ──────────────────────────────────────────────────────────────
BASE_DIR = Path("/home/preonath/Desktop/Preonath_Project/Zika_Dengue")
os.chdir(BASE_DIR)   # all relative paths resolve from here
print(f"Working directory: {Path.cwd()}")

# ── Verify conda environment ──────────────────────────────────────────────────
print(f"Python: {sys.executable}")
try:
    import scanpy, pydeseq2, gseapy, mygene, GEOparse, networkx
    print("✓ All required packages found")
except ImportError as e:
    print(f"✗ Missing package: {e}")
    print("  Activate conda env:  conda activate scipy")

# ── Step runner: loads script and calls main() ─────────────────────────────────
# Each script has its own checkpoint logic — completed steps are skipped automatically.
def run_step(script_filename: str):
    """Load a pipeline script and call its main() function."""
    script_path = BASE_DIR / "scripts" / script_filename
    print(f"\n{'='*60}")
    print(f"RUNNING: {script_filename}")
    print(f"{'='*60}")
    spec = importlib.util.spec_from_file_location(
        script_filename.replace(".py", ""), str(script_path)
    )
    mod = importlib.util.module_from_spec(spec)
    spec.loader.exec_module(mod)
    if hasattr(mod, "main"):
        mod.main()
    else:
        print(f"  WARNING: {script_filename} has no main() function")
    return mod

# ── Checkpoint viewer ─────────────────────────────────────────────────────────
def show_checkpoint(step_num: int):
    """Print current checkpoint state for a step."""
    ckpt_files = {
        1: "step01_checkpoint.json",  2: "step02_checkpoint.json",
        3: "step03_checkpoint.json",  4: "step04_checkpoint.json",
        5: "step05_checkpoint.json",  6: "step06_checkpoint.json",
        7: "step07_checkpoint.json",  8: "step08_checkpoint.json",
        9: "step09_checkpoint.json",  10: "step10_checkpoint.json",
        11: "step11_checkpoint.json", 12: "step12_checkpoint.json",
        13: "step13_checkpoint.json", 14: "step14_checkpoint.json",
        15: "step15_checkpoint.json",
    }
    fp = BASE_DIR / "checkpoints" / ckpt_files.get(step_num, "")
    if fp.exists():
        ckpt = json.load(open(fp))
        print(f"Step {step_num:02d} checkpoint: {json.dumps(ckpt, indent=2)}")
    else:
        print(f"Step {step_num:02d}: No checkpoint found (not yet run)")

print("\n✓ Setup complete. Ready to run pipeline.")
print("  To run all: execute each cell below in order, or use Kernel → Restart & Run All")
````

**Key result:** 2,260 → 2,075 cells (8% removed — well within <30% SOP threshold)

> **Note on MT%:** pct_counts_mt = 0 for all cells in this dataset.  
> The Zanini lab used non-standard MT gene naming — MT filtering had no effect.

## PHASE 2 — Pseudobulk Differential Expression
### Step 03: Pseudobulk DESeq2 → Shared DEGs → FC Correlation

**What it does:**
- Aggregates raw counts per condition × timepoint (12 pseudobulk samples total)
- Runs **pydeseq2** (Python port of DESeq2, ~condition + timepoint model) for:
  - DENV vs Control
  - ZIKV vs Control
- Threshold: |log2FC| ≥ 1, padj < 0.05
- Finds shared upregulated and downregulated DEGs
- Fisher exact test for shared overlap enrichment → **GATE G2**
- Pearson correlation of genome-wide FC (DENV vs ZIKV) → **GATE G3**

**Key results (confirmed from checkpoints):**
- DENV DEGs: **527** (300 up, 227 down)
- ZIKV DEGs: **176** (66 up, 110 down)
- **15 shared upregulated genes** (BIRC3, CCL4, CREBRF, CXCL1, DUSP1, INHBE, LPXN, PLA2G4C, RND1, SIRT4, TSPAN1, TSPYL2, VNN3P, CFAP251, CD200R1)
- GATE G2: Fisher p = 2.26×10⁻¹⁵, Fold Enrichment = 18.56× ✅ **PASSED**
- GATE G3: Pearson r = 0.357 overall; r = 0.462 at 48h ⚠️ **MODERATE / PASS at 48h**

**Checkpoint:** `checkpoints/step03_checkpoint.json`

````python
run_step("step03_pseudobulk_deg.py")
show_checkpoint(3)
````

````python
# Verify DEG results
import pandas as pd
from pathlib import Path

DEG_DIR = BASE_DIR / "01_processed_data/deg_tables"
RES_DIR = BASE_DIR / "03_results/phase3_shared_degs"

denv = pd.read_csv(DEG_DIR / "DEGs_DENV_vs_Control.csv")
zikv = pd.read_csv(DEG_DIR / "DEGs_ZIKV_vs_Control.csv")
shared = pd.read_csv(RES_DIR / "shared_DEGs_shared_up.csv")

denv_sig = denv[(denv['padj']<0.05) & (denv['log2FoldChange'].abs()>=1)]
zikv_sig = zikv[(zikv['padj']<0.05) & (zikv['log2FoldChange'].abs()>=1)]

print(f"DENV DEGs: {(denv_sig['log2FoldChange']>0).sum()} up / {(denv_sig['log2FoldChange']<0).sum()} down")
print(f"ZIKV DEGs: {(zikv_sig['log2FoldChange']>0).sum()} up / {(zikv_sig['log2FoldChange']<0).sum()} down")
print(f"Shared upregulated: {len(shared)} genes")

corr = pd.read_csv(RES_DIR / "correlation_results.csv")
print(f"\nPearson r (genome-wide FC): {corr['Pearson_r'].values[0]:.4f}  p={corr['Pearson_p'].values[0]:.2e}")
````

## Known Limitations (for Methods section)

1. **Tool mismatch** — SOP specified R/DESeq2; pipeline used Python/pydeseq2 (statistically equivalent).
   Add to Methods: *"Python equivalents were used in place of R packages; pydeseq2 implements the same negative binomial GLM as DESeq2."*

2. **MT% = 0** — Mitochondrial genes were not detected (dataset uses non-standard gene naming).
   MT filtering had no effect. Note in Methods.

3. **ZIKV G1 borderline** — 176 DEGs vs ≥200 SOP threshold.
   ZIKV has fewer cells per timepoint than DENV in this dataset.

4. **Gate G5 not FDR significant** — CREBRF hub is nominally significant (small n=15 limits power).

5. **CFAP251** — Ciliary protein in hepatocyte cell line (Huh7 cells have no cilia). Note in Discussion.

6. **VNN3P** — Pseudogene. Note in Discussion; exclude from protein-level interpretation.
